```python
import os
import time
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.image import img_to_graph
import xarray as xr
import netCDF4 as nc
import gzip
```

### Parameters

```python
# datasets path
PATH_PRECIP = './data/precipitations/'
DATA_PATH = './data/precipitations/daily_precip_1degree_amazone20x20_1982-2019.npy'

# parameters related to datasets
LEN_TIME = 31
SYSTEM_DIM = np.array([20,20])  # dimension of input
```

```python
ds = []
for filename in os.listdir(PATH_PRECIP):
    f = os.path.join(PATH_PRECIP, filename)
    # checking if it is a file
    if os.path.isfile(f):
        with gzip.open(f) as gz_file:
            ds.append(nc.Dataset('dummy', mode='r', memory=gz_file.read()))
```

```python
fig, ax = plt.subplots(figsize=(18, 4))
ax.imshow(np.flip(ds[0]['precip'], axis=(0,1))[25,:,:])
plt.tight_layout()
plt.axis('off');
```

```python
raw_ds = np.concatenate([np.flip(d['precip'], axis=(0,1)) for d in ds]) # concatenate all 38 years of data
```

```python
raw_ds.shape # 13879 = 38 years * 365 days + 9 (leap years)
```

```python
mask = np.where(raw_ds[0].data==-9999., False, True)
```

```python
#graphs = [img_to_graph(img=im, mask=mask) for im in raw_ds]
adj = [img_to_graph(img=snapshot.data, mask=mask) for snapshot in raw_ds]
adj[0].shape
```

```python
adj.data[adj.data>0.] = 1 # binary adjadency matrix
(adj.data-adj.data.T).sum() # check symmetry
```

```python
features = raw_ds[0].data[mask] # define mm of precipitation as nodes featrues
features.shape
```

```python
with open('adj.pkl', 'wb') as f:
    pickle.dump(adj, f)
```
